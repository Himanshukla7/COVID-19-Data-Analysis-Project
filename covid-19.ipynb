```python
def country_wise_data(data, country_name):
    data_country_name = data[data['Country/Region'] == country_name].reset_index(drop=True)
    data_country_name.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=1, inplace=True)
    data_country_name = data_country_name.T
    return data_country_name
```

```python
%matplotlib inline

def covid_visulization(data, countries = []):
    
    plt.figure(figsize=(8,6))
    
    for country in countries:
        
        data_country = country_wise_data(data, country_name=country)
        
        data_country.index = pd.DatetimeIndex(data_country.index)
        
        number_of_cases = data_country.values
        
        dates = data_country.index.values
        
        plt.plot(dates, number_of_cases, label=country)
        
    plt.legend()
    
    plt.grid()
    
    plt.show()
        
```

```python
covid_visulization(data, countries = ["India", "Afghanistan", "Italy", "Germany"])
```

```python
### Q1: Which country had highest number of cases till the month of April, 2020

### Answer: US
```

```python
total_cases = data.iloc[:, 4:].sum(axis=1)

countries = data["Country/Region"]

country_wise_total_cases = pd.concat([countries, total_cases], axis = 1).rename(columns={"Country/Region": "Country", 0:"total_cases"})
```

```python
country_wise_total_cases
```

```python
max_count = country_wise_total_cases['total_cases'].max()

country_wise_total_cases[country_wise_total_cases['total_cases'] == max_count]
```

```python
min_count = country_wise_total_cases['total_cases'].min()

country_wise_total_cases[country_wise_total_cases['total_cases'] == min_count]
```
